```python
iris_setosa = iris.loc[iris['Species'] == 'Iris-setosa']
iris_virginica = iris.loc[iris['Species'] == 'Iris-virginica']
iris_versicolor = iris.loc[iris['Species'] == 'Iris-versicolor']
```

```python
plt.plot(iris_setosa['Petal_Length(cm)'], np.zeros_like(iris_setosa['Petal_Length(cm)']), 'o')
plt.plot(iris_virginica['Petal_Length(cm)'], np.zeros_like(iris_virginica['Petal_Length(cm)']), 'o')
plt.plot(iris_versicolor['Petal_Length(cm)'], np.zeros_like(iris_versicolor['Petal_Length(cm)']), 'o')
plt.show()
```

```python
# Tutorial no. 5
counts1, bin_edges1 = np.histogram(iris_setosa['Petal_Length(cm)'], bins=10, density=True)
counts2, bin_edges2 = np.histogram(iris_virginica['Petal_Length(cm)'], bins=10, density=True)
counts3, bin_edges3 = np.histogram(iris_versicolor['Petal_Length(cm)'], bins=10, density=True)

pdf1 = counts1/sum(counts1)
pdf2 = counts2/sum(counts2)
pdf3 = counts3/sum(counts3)

print(pdf1, pdf2, pdf3)
print(bin_edges1, bin_edges2, bin_edges3)
```

```python
# Computing CDF
cdf1 = np.cumsum(pdf1)
cdf2 = np.cumsum(pdf2)
cdf3 = np.cumsum(pdf3)

cdf1, cdf2, cdf3
```

```python
# Plotting Graphs
plt.plot(bin_edges1[1:], pdf1)
plt.plot(bin_edges2[1:], pdf2)
plt.plot(bin_edges3[1:], pdf3)
plt.plot(bin_edges1[1:], cdf1)
plt.plot(bin_edges2[1:], cdf2)
plt.plot(bin_edges3[1:], cdf3)
plt.xlabel('Petal_Length(cm)')
plt.ylabel('Probability')
plt.show()
```

```python
# Tutorial no. 6
# Mean and Standard Deviation
print('Mean')
print(np.mean(iris_setosa['Petal_Length(cm)']))
print(np.mean(iris_virginica['Petal_Length(cm)']))
print(np.mean(iris_versicolor['Petal_Length(cm)']))

print('\nStandard Deviation')
print(np.std(iris_setosa['Petal_Length(cm)']))
print(np.std(iris_virginica['Petal_Length(cm)']))
print(np.std(iris_versicolor['Petal_Length(cm)']))
```

```python
# Tutorial no. 7
# Mean absolute deviation
from statsmodels import robust
print('Median absolute Deviation')
print(robust.mad(iris_setosa['Petal_Length(cm)']))
print(robust.mad(iris_virginica['Petal_Length(cm)']))
print(robust.mad(iris_versicolor['Petal_Length(cm)']))
```

```python
# Tutorial no. 8
# Haberman's cancer survival dataset analysis
import pandas as pd
import seaborn as snb
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('haberman.csv')
df
```

```python
df.columns = ['Age', 'Operation_Year', 'Auxillary_Lymph_Node', 'Survival_Status']
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df['Survival_Status'] = df['Survival_Status'].map({1:'Yes', 2:'No'})
```

```python
df
```

```python
df['Survival_Status'].value_counts()
```

```python
df.describe()
```

```python
Yes_Status = df[df['Survival_Status'] == 'Yes']
Yes_Status.describe()
```

```python
No_Status = df[df['Survival_Status'] == 'No']
No_Status.describe()
```

```python
sns.FacetGrid(df, hue='Survival_Status', height = 5).map(sns.distplot, 'Age').add_legend()
sns.FacetGrid(df, hue='Survival_Status', height = 5).map(sns.distplot, 'Operation_Year').add_legend()
sns.FacetGrid(df, hue='Survival_Status', height = 5).map(sns.distplot, 'Auxillary_Lymph_Node').add_legend()
plt.show()
```

```python
# CDF

counts1, bin_edges1 = np.histogram(Yes_Status['Auxillary_Lymph_Node'], bins=10, density=True)
counts2, bin_edges2 = np.histogram(No_Status['Auxillary_Lymph_Node'], bins=10, density=True)

pdf1 = counts1/(sum(counts1))
pdf2 = counts2/(sum(counts2))

cdf1 = np.cumsum(pdf1)
cdf2 = np.cumsum(pdf2)

plt.plot(bin_edges1[1:], pdf1)
plt.plot(bin_edges1[1:], cdf1, label='Yes')

plt.plot(bin_edges2[1:], pdf2)
plt.plot(bin_edges2[1:], cdf2, label='No')
plt.xlabel('Nodes')
plt.legend()
plt.show()
```

```python
sns.boxplot(x='Survival_Status', y='Age', data=df)
plt.show()
sns.boxplot(x='Survival_Status', y='Operation_Year', data=df)
plt.show()
sns.boxplot(x='Survival_Status', y='Auxillary_Lymph_Node', data=df)
plt.show()
```

```python
sns.violinplot(x='Survival_Status', y='Age', data=df)
plt.show()
sns.violinplot(x='Survival_Status', y='Operation_Year', data=df)
plt.show()
sns.violinplot(x='Survival_Status', y='Auxillary_Lymph_Node', data=df)
plt.show()
```

```python
# Tutorial no. 35
from scipy import stats
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5, size=500) + 5
prob = stats.probplot(x, dist=stats.norm, plot=ax1)
ax1.set_xlabel("")
ax1.set_title("ProbPlot against normal distribution")
ax2 = fig.add_subplot(212)
xt, _ = stats.boxcox(x)
prob = stats.probplot(xt, dist = stats.norm, plot=ax2)
ax2.set_title("ProbPlot after box-cox transformation")
plt.show()
```
